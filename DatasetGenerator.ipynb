```python
import numpy as np
import pandas as pd
import csv
import os
from faker import Faker
fake = Faker()

def create_line(number_of_columns):
    part_1 = int(number_of_columns * 5/100) # 5
    part_2 = int(part_1 + number_of_columns * 5/100)
    part_3 = int(part_2 + number_of_columns * 5/100)
    part_4 = int(part_3 + number_of_columns * 85/100 / 2)

    line = []
    for j in range(0, number_of_columns):
        if j < part_1:
            line.append(fake.boolean())
        
        elif part_1 <= j < part_2:
            line.append(fake.date())
        
        elif part_2 <= j < part_3:
            line.append(fake.paragraph(nb_sentences=1))
        
        elif part_3 <= j < part_4:
            line.append(fake.random_int())
        
        else:
            line.append(fake.pyfloat())
    
    return line


def create_data_set_file(number_of_lines, number_of_columns, filename):
    final_array = []

    with open(filename, 'w', newline='') as file:
            writer = csv.writer(file)

    part_1 = int(number_of_columns * 5/100) # 5
    part_2 = int(part_1 + number_of_columns * 5/100)
    part_3 = int(part_2 + number_of_columns * 5/100)
    part_4 = int(part_3 + number_of_columns * 85/100 / 2)
    print(part_1, ' - ', part_2, ' - ', part_3, ' - ', part_4, "CSV file generated!")
    
    for i in range(0, number_of_lines):
        line = create_line(number_of_columns)

        final_array.append(line)
        with open(filename, 'a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(line)

    #print(final_array)


def main():

    compteur = 1
    number_of_lines = 100
    number_of_columns = 100
    while True:
        filename = 'test' + str(compteur) + '.csv'

        create_data_set_file(number_of_lines, number_of_columns, filename)
            
        file_size_in_bytes = os.path.getsize('/Users/davidta/Desktop/'+filename)
        file_size_in_mb = file_size_in_bytes / 1000000
        if file_size_in_mb > 49:
            break

        number_of_lines += 0
        number_of_columns += 1000
        compteur +=1
        print (filename)
main()
```
